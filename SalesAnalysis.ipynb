```python
import os
import pandas as pd
```

###  <font color='blue'>Reading a Single File</font>

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
```

###  <font color='blue'>Getting Names of all Files in the Sales Data Folder</font>

```python
files = [file for file in os.listdir("./Sales_Data")]
files
```

###  <font color='blue'>Concatenating all Files into a single Data Frame</font>

```python
df = pd.DataFrame()
for file in files:
    df = pd.concat([df,pd.read_csv("./Sales_Data/"+file)])
df
```

###  <font color='blue'>Creating a new column "Month"</font>

```python
df['Month'] = df['Order Date'].str[:2]
df
```

###  <font color='blue'>Dropping Null Value Rows</font>

```python
df=df.dropna(how="all")
```

```python
df
```

###  <font color='blue'>Month Column still not clean as there are Rows with Values as "Or"</font>

###  <font color='blue'>Changing "Quantity Ordered" column and "Price Each" column to Numeric</font>

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df.dtypes
```

###  <font color='blue'>Creating new column "Sales" as the product of "Price Each" and "Quantity Ordered"</font>

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

###  <font color='blue'>Extracting City and State from "Purchase Address"</font>

```python
def get_city(address):
    return address.split(", ")[1]
def get_state(address):
    return address.split(", ")[2][:2]

df['City'] = df['Purchase Address'].apply(lambda address: f"{get_city(address)} ({get_state(address)})")
df
```

###  <font color='blue'>Grouping Data by City Name</font>

```python
city_results = df.groupby(['City']).sum()
city_results
```

###  <font color='blue'>Creating a List for storing Unique City Names</font>

```python
cities = [city for city, d in df.groupby(['City'])]
cities
```

###  <font color='blue'>Plotting Total Sales against City Name</font>

```python
import matplotlib.pyplot as plt
plt.bar(cities,city_results['Sales'])

plt.xlabel("City Name")
plt.ylabel("Sales in USD$")
plt.xticks(cities,rotation = 60)
plt.show()
```

###  <font color='blue'>Keeping only rows with duplicated Order IDs</font>

```python
df['Order ID'].duplicated(keep = False)
```

```python
duplicated_df = df.loc[df['Order ID'].duplicated(keep = False)]
duplicated_df.head()
```

###  <font color='blue'>Making a new column "Products" which stores all the products ordered in a single order</font>

```python
duplicated_df['Products'] = duplicated_df.groupby('Order ID')['Product'].transform(lambda x: ", ".join(x))
duplicated_df.head()
```

###  <font color='blue'>Dropping rows with Duplicated Order IDs as now all the products from a single order is stored in the column "Products"</font>

```python
duplicated_df = duplicated_df[['Order ID', 'Products']].drop_duplicates()
duplicated_df.head()
```

###  <font color='blue'>Getting the top 10 products which are bought the most in pairs</font>

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in duplicated_df['Products']:
    rows = row.split(", ")
    count.update(Counter(combinations(rows,2)))
    
for key,value in count.most_common(10):
    print(f"{key}: {value}")
```

###  <font color='blue'>Getting the top 10 products which are bought the most in a group of three</font>

```python
count = Counter()
for row in duplicated_df['Products']:
    rows = row.split(", ")
    count.update(Counter(combinations(rows,3)))
    
for key,value in count.most_common(10):
    print(f"{key}: {value}")
```

###  <font color='blue'>Grouping the data by Product</font>

```python
products_df = df.groupby('Product')
qty = products_df.sum()['Quantity Ordered']
products = [product for product, df in products_df]
```

###  <font color='blue'>Plotting Quantity Ordered against Product Names</font>

```python
plt.bar(products,qty)

plt.xlabel("Product Names")
plt.ylabel("Quantity Ordered")
plt.xticks(products,rotation = 90)
plt.show()
```

###  <font color='blue'>Getting the price of each product</font>

```python
prices = products_df.mean()['Price Each']
prices
```

###  <font color='blue'>Plotting Quantity Ordered and Prices against Product Names</font>

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.bar(products,qty,color='green')
ax2.plot(products,prices,color='blue')

ax1.set_xlabel("Product Names")
ax1.set_ylabel("Quantity Ordered")
ax2.set_ylabel("Prices")

ax1.set_xticklabels(products, rotation = 90)

plt.show()
```
